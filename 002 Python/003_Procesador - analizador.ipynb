```python
class ProcesadorTexto: #Esta es la clase, por lo tanto es como si dijera "tengo un plano" con ese plano voy a poder construir mis objetos, el objeto que yo quiero es un analizador
    """
    Una clase simple que encapsula un texto y ofrece
    herramientas básicas para analizarlo.
    """

    # El "constructor" se ejecuta al crear un nuevo objeto.
    # Define los atributos iniciales.
    def __init__(self, texto_a_procesar): #El init es el que iniciliza la clase, y tiene un parametro self que llama a la clase a si misma y tiene un agumento, el texto a procesar
        print(f"Objeto ProcesadorTexto creado con éxito para el texto.")
        # Atributo 1: guardamos el texto original.
        self.texto_original = texto_a_procesar
        # Atributo 2: creamos una versión limpia para trabajar.
        self.palabras = self.texto_original.lower().split() #convertir en minisculas y split para que cree una lista con cada palabra

    # Método 1: una acción que el objeto puede realizar. Contruyo funcion para contar palabra
    def contar_palabras(self):
        """Devuelve el número total de palabras."""
        return len(self.palabras)

    # Método 2: otra acción disponible.
    def obtener_palabras_unicas(self):
        """Devuelve una lista de palabras sin repeticiones."""
        # "set" elimina duplicados automáticamente.
        return list(set(self.palabras)) #El metodo set elimina duplicados de palabras
```

```python
# 1. IMPORTACIÓN: Traemos el plano (la clase) desde nuestro otro archivo.
# Le decimos a Python: "Andá al archivo 'procesador' y traéme la herramienta 'ProcesadorTexto'."
#from procesador import ProcesadorTexto
#solamente tiene sentido cuando tu clase ProcesadorTexto está en otro archivo llamado procesador.py (en la misma carpeta o en el PYTHONPATH). Es decir:
#procesador.py → contiene la clase ProcesadorTexto
#analizador.py → importa la clase desde ese archivo con from procesador import ProcesadorTexto
#Esto se usa para llamar a las librerias

print("Iniciando programa analizador")
print("-" * 20)

# El texto que queremos analizar.
mi_texto = "La Programación Orientada a Objetos es más intuitiva de lo que parece."

# 2. CREACIÓN DEL OBJETO (INSTANCIACIÓN):
# Usamos la clase importada como un molde para crear un objeto real y funcional.
# Acá es donde el "plano" se convierte en una "casa".
analizador_de_texto = ProcesadorTexto(mi_texto)

print("-" * 20)

# 3. USO DE SUS MÉTODOS:
# Ahora que tenemos nuestro objeto 'analizador_de_texto', podemos usar las
# habilidades (métodos) que definimos en su plano.
total_palabras = analizador_de_texto.contar_palabras()
palabras_unicas = analizador_de_texto.obtener_palabras_unicas()

print(f"El texto tiene {total_palabras} palabras en total.")
print(f"Las palabras únicas son: {palabras_unicas}")
```
