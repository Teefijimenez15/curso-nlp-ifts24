```python
import gradio as gr #Importamos la librería para usarla en el código.
```

```python
def bienvenida(name):
    return "Hola " + name #Definimos una función simple que recibe un nombre (name) y devuelve el saludo "Hola <nombre>"


bienvenido = gr.Interface(fn=bienvenida, inputs="text", outputs="text") #Creamos una interfaz gráfica con Gradio. #fn=bienvenida → le decimos qué función queremos conectar a la interfaz. inputs="text" → la entrada será una caja de texto donde el usuario escribirá su nombre. outputs="text" → la salida será texto (el saludo que devuelve la función).

bienvenido.launch() #Lanza la interfaz en el navegador. Se abre un servidor local y te da un link donde podés interactuar con la app. Al escribir un nombre (ejemplo: Stefania), la interfaz devuelve: Hola Stefania
```

```python
def saludo(nombre, intensidad): # Define la función 'saludo' con dos parámetros:
                                 #  - nombre: string con el nombre a saludar
                                 #  - intensidad: número (o string convertible a int) que indica cuántos "!" usar
  return "Hola, " + nombre + "!" * int(intensidad)  # Construye el saludo concatenando:
                                                      #  "Hola, " + nombre + ("!" repetido 'intensidad' veces).
                                                      #  Nota: el operador * tiene mayor precedencia que +,
                                                      #  por eso es equivalente a: "Hola, " + nombre + ("!" * int(intensidad))
```

```python
demo = gr.Interface( ## Creamos una interfaz con Gradio y la guardamos en la variable 'demo'
  fn=saludo, #  'fn' es la función de Python que vamos a conectar a la interfaz (en este caso 'saludo')
  inputs=["text", "slider"], # 'inputs' define qué elementos de entrada tendrá la interfaz:
                               #   - "text": un cuadro de texto donde el usuario escribe su nombre
                               #   - "slider": un control deslizante para elegir la intensidad (número de "!")
  outputs=["text"],
)

demo.launch() # Lanza la aplicación web de Gradio en el navegador para interactuar con la interfaz
```

```python
def greet(name):
    return "Hello " + name


# We instantiate the Textbox class
textbox = gr.Textbox(label="Type your name here:", placeholder="John Doe", lines=2)

gr.Interface(fn=greet, inputs=textbox, outputs="text").launch()
```

```python
from bs4 import BeautifulSoup #es una librería que sirve para parsear HTML y XML (o sea, leer el código de una página web y navegarlo como si fuera un árbol).
```

```python
html = """
<html>
<head>
  <title>Mi Página de Ejemplo</title>
</head>
<body>
  <h1>Título Principal</h1>
  <p>Este es un párrafo.</p>
  <a href="https://www.ejemplo.com">Enlace a Ejemplo</a>
</body>
</html>
""" #cargo el HTML
```

```python
soup = BeautifulSoup(html, 'html.parser') # Aquí se hace el parsing, detecta todas las etiquetas # Crea un objeto BeautifulSoup a partir del contenido HTML
                                            # - 'html' es la variable que contiene el código fuente de la página
                                            # - 'html.parser' indica que se usará el parser de Python para analizar el HTML
                                            # El resultado es un objeto que permite recorrer y extraer datos del documento

print(soup)       # Imprime todo el objeto BeautifulSoup, es decir,
                                            # muestra el HTML ya "parseado" y formateado
                                            # Útil para verificar que se haya cargado bien la estructura
```

```python
titulo = soup.title   # Accede directamente a la primera etiqueta <title> del documento HTML
print(titulo)       # Imprime la etiqueta <title> Accede al titulo
print(titulo.text)  # Imprime el texto dentro de la etiqueta <title>, imprime texto plano
```

```python
parrafo = soup.p #etiqueta P
print(parrafo)      # Imprime la etiqueta <p>
print(parrafo.text) # Imprime el texto dentro de la etiqueta <p>
```

```python
html = """
<html><head><title>Mi Página de Ejemplo</title></head><body><h1>Título Principal</h1><p>Este es un párrafo.</p><a href="https://www.ejemplo.com">Enlace a Ejemplo</a></body></html>
"""

soup = BeautifulSoup(html, 'html.parser')

print(soup.prettify())
```

```python
html = """
<a href="https://www.ejemplo.com" class="enlace-principal">Enlace a Ejemplo</a>
"""

soup = BeautifulSoup(html, 'html.parser')

enlace = soup.a
print(enlace['href'])   # Imprime el valor del atributo 'href'
print(enlace['class'])  # Imprime el valor del atributo 'class'
```
