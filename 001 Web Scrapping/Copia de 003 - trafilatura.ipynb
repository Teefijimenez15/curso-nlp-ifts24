# ¬øQu√© es Trafilatura?

Es una librer√≠a de Python dise√±ada para extraer autom√°ticamente el texto principal de art√≠culos en p√°ginas web.

A diferencia de BeautifulSoup (donde vos ten√©s que buscar manualmente los p o div), Trafilatura ya est√° optimizada para detectar y devolver solo el contenido central de un art√≠culo, sin men√∫s, publicidad, ni pie de p√°gina.

üëâ Ideal para construir corpus de textos period√≠sticos, blogs, revistas.

## ¬øQu√© es un corpus de textos?

Un corpus es una colecci√≥n de textos organizados que sirven como base para an√°lisis, miner√≠a de texto o entrenamiento de modelos de lenguaje.

Ejemplo: pod√©s scrapear notas de Clar√≠n, P√°gina/12 o Revista √ë, sobre un tema (ej: pol√≠tica, econom√≠a, cultura, un personaje famoso).

Ese corpus luego se convierte en un dataset.

##¬øQu√© es metadata?

Metadata = datos sobre los datos.

En este caso, no solo quer√©s el texto del art√≠culo, sino tambi√©n informaci√≥n adicional:
URL de origen
Fecha de publicaci√≥n
Autor
Medio (P√°gina/12, Clar√≠n, etc.)
Tema o categor√≠a
La metadata es clave para:
Recuperaci√≥n aumentada de datos (ej. RAG con modelos de lenguaje).
Filtrar corpus por fecha, fuente o tem√°tica.
Mantener trazabilidad y transparencia del dataset.

##Explicaci√≥n del codigo
El c√≥digo usa la librer√≠a Trafilatura para descargar y extraer el texto principal de art√≠culos de una p√°gina web, en este caso una nota de P√°gina/12. Primero, se define una funci√≥n que recibe una URL y trata de obtener el contenido de la p√°gina. Si logra descargarlo, intenta extraer el texto limpio (sin comentarios ni im√°genes). Si todo sale bien, devuelve ese texto; si no, devuelve un mensaje de error. Esto evita que el programa se corte cuando hay problemas con alguna URL.

Luego, se prepara una lista de URLs (pod√©s agregar tantas como quieras) y se crea una lista vac√≠a para ir guardando los datos. Por cada URL, el programa muestra en consola cu√°l est√° procesando, llama a la funci√≥n para obtener el texto y lo guarda junto con la URL en un diccionario. Ese diccionario se va agregando a la lista corpus_data.

Despu√©s, con pandas, se transforma esa lista en un DataFrame (una tabla) donde cada fila corresponde a un art√≠culo y las columnas son la URL y el texto extra√≠do. As√≠ se construye un corpus de textos period√≠sticos.

El programa tambi√©n imprime las primeras filas de esa tabla para revisarla, y recorre todas las filas mostrando por consola el texto de cada art√≠culo con su URL correspondiente. Por √∫ltimo, permite acceder de forma directa al texto de un art√≠culo en particular (por ejemplo, el primero de la lista).

```python
import pandas as pd
import trafilatura

def obtener_texto_articulo(url):
    """
    Obtiene y extrae texto de una URL dada usando trafilatura.

    Args:
        url (str): La URL del art√≠culo.

    Returns:
        str: El texto extra√≠do del art√≠culo, o un mensaje de error si la obtenci√≥n o extracci√≥n falla.
    """
    try:
        descargado = trafilatura.fetch_url(url)
        if descargado:
            texto = trafilatura.extract(descargado, include_comments=False, include_images=False)
            if texto:
                return texto
            else:
                return "No se pudo extraer texto de la URL."
        else:
            return "No se pudo obtener la URL."
    except Exception as e:
        return f"Ocurri√≥ un error: {e}"

# Lista de URLs para construir el corpus (puedes modificar esta lista)
urls = [
    "https://www.pagina12.com.ar/849012-taylor-swift-anuncia-el-lanzamiento-de-su-nuevo-disco-the-li",
    # Agrega m√°s URLs aqu√≠ si es necesario
]

# Crear una lista para almacenar los datos
corpus_data = []

# Iterar sobre las URLs y extraer el texto
for url in urls:
    print(f"Procesando: {url}")
    texto_extraido = obtener_texto_articulo(url)
    corpus_data.append({'URL': url, 'Texto': texto_extraido})

# Crear el DataFrame de pandas
df_corpus = pd.DataFrame(corpus_data)

# Mostrar las primeras filas del DataFrame
display(df_corpus.head())
```

```python
# Iterar sobre las filas del DataFrame e imprimir el texto completo
for index, row in df_corpus.iterrows():
    print(f"--- Texto del art√≠culo de la URL: {row['URL']} ---")
    print(row['Texto'])
    print("-" * 30) # Separador entre art√≠culos
```

```python
# Tambi√©n podes acceder al texto de un art√≠culo espec√≠fico por su √≠ndice
# Por ejemplo, para el primer art√≠culo (√≠ndice 0):
print(df_corpus.loc[0, 'Texto'])
```
